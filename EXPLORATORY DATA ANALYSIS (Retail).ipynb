### STEP 3 - Data Manipulation 

```python
# Dropping irrelevant columns
data1 = data.drop(['Country','Postal Code'],axis=1)
data1.sample(10)
```

```python
# Checking for duplicate rows
data1.duplicated().sum()
```

```python
# Dropping the duplicates 
data1.drop_duplicates()
```

```python
#STEP 6 Dropping the missing or null values.
data1.isnull().sum()
```

### STEP 3 - Data Visualization 

```python
sns.heatmap(data1.corr(),annot= True)
```

Sales and Profit are Moderately Correlated.

Discount and Profit are Negatively Correlated

Quantity and Profit are less Moderately Correlated.

### Plotting Graphs with respect to Orders

#### Plotting Orders Vs Category

```python
sns.countplot(data1.Category)
```

Most of the orders are from the Office Supplies Category. (We can increase the prices of these to earn more.)

The orders from Furniture and Technology Categories are almost same and a lot less than that of Office Supplies.

#### Plotting Orders Vs Segment

```python
sns.countplot(data1.Segment)
```

Consumer Segment has the most no. of orders, followed by Corporate Segment and then the Home Office.

#### Plotting Orders Vs Region 

```python
sns.countplot(data1.Region)
```

#### Plotting Orders Vs Ship Mode 

```python
sns.countplot(data1['Ship Mode'])
```

#### Plotting Orders Vs Sub-Category

```python
sns.countplot(data1['Sub-Category'])
plt.rcParams["figure.figsize"] = (21,4)
```

#### Plotting Orders Vs State

```python
plt.figure(figsize=(20,6))
sns.countplot(x=data1['State'], order=data1['State'].value_counts().index)
plt.xlabel('STATE')
plt.ylabel('COUNT')
plt.xticks(rotation=90)
```

### Plotting Graphs with respect to Sales

#### Category Vs Sales

```python
plt.figure(figsize=(10,4))
plt.bar('Category', 'Sales', data=data1)
plt.title('CATEGORY VS SALES GRAPH')
plt.xlabel("CATEGORY")
plt.ylabel("SALES")
```

We can see that the sales of Technology Category is the highest but its orders are the least(Shown in the above orders vs category graph). 

#### Sub-Category Vs Sales

```python
plt.figure(figsize=(20,4))
plt.bar('Sub-Category', 'Sales', data=data1)
plt.title('SUB-CATEGORY VS SALES GRAPH')
plt.xlabel("SUB-CATEGORY")
plt.ylabel("SALES")
```

#### State Vs Sales

```python
plt.figure(figsize=(20,4))
plt.bar('State', 'Sales', data=data1)
plt.xticks(rotation=90, fontsize=10)
plt.title('STATE VS SALES GRAPH')
plt.xlabel("STATE")
plt.ylabel("SALES")
```

### Plotting Graphs with respect to Profit

#### Category Vs Profit

```python
plt.figure(figsize=(10,4))
plt.bar('Category', 'Profit', data=data1, color='green')
plt.title('CATEGORY VS PROFIT GRAPH')
plt.xlabel("CATEGORY")
plt.ylabel("PROFIT")
```

#### Sub-Category Vs Profit

```python
plt.figure(figsize=(20,4))
plt.bar('Sub-Category', 'Profit', data=data1, color='cyan')
plt.title('SUB-CATEGORY VS PROFIT GRAPH')
plt.xlabel("SUB-CATEGORY")
plt.ylabel("PROFIT")
```

#### State Vs Profit

```python
plt.figure(figsize=(20,10))
plt.bar('State', 'Profit', data=data1, color='red')
plt.xticks(rotation=90, fontsize=10)
plt.title('STATE VS PROFIT GRAPH')
plt.xlabel("STATE")
plt.ylabel("PROFIT")
```

#### Plotting Profit Vs Discount line plot

```python
sns.lineplot(data1.Discount, data1.Profit, color='black')
```

#### Plotting Quantity Vs Profit line plot

```python
sns.lineplot(data1.Quantity, data1.Profit, color='black')
```
